Эмпирическая функция:
Эмперической функцией распределения называется функция $\tilde{F}(x)=\frac{m(x)}{n}$,  
где $m(x)$ - число наблюдений, попавших в интервал $(-\infty, x)$  
$\begin{cases}-\infty < x \leq 7.95 & 0 \\ 7.95 <x \leq 8.02 & \frac{1}{12} \\ 8.02 < x \leq 8.08 & \frac{2}{12} \\ 8.08 < x \leq 8.09 & \frac{3}{12} \\ 8.09 < x \leq 8.16 & \frac{4}{12} \\ 8.16 < x \leq 8.21 & \frac{5}{12} \\ 8.21 < x \leq 8.23& \frac{6}{12} \\ 8.23 < x \leq 8.27 & \frac{7}{12} \\ 8.27 < x \leq 8.32 & \frac{8}{12} \\ 8.32 < x \leq 8.33 & \frac{9}{12} \\ 8.33 < x \leq 8.36 & \frac{10}{12} \\ 8.36 < x \leq 8.42 & \frac{11}{12} \\8.42 < x < \infty & \frac{12}{12}\end{cases}$

```python
# Данные
sample = [7.95, 8.02, 8.08, 8.09, 8.16, 8.21, 8.23, 8.27, 8.32, 8.33, 8.36, 8.42]
n = len(sample)
mean = 8.2   # Среднее
sigma = 0.1  # Стандартное отклонение

# Эмпирическая функция распределения (ЭФР)
empirical_x = []
empirical_y = []

for i in range(n):
    empirical_x.extend([sample[i], sample[i]])
    empirical_y.extend([i / n, (i + 1) / n])

# Добавим точки в начале и в конце для замыкания ЭФР
empirical_x = [min(sample) - 0.1] + empirical_x + [max(sample) + 0.1]
empirical_y = [0] + empirical_y + [1]

# Теоретическая функция распределения (нормальное распределение)
import math

def normal_cdf(x, mean, sigma):
    z = (x - mean) / sigma
    return 0.5 * (1 + math.erf(z / math.sqrt(2)))

x_vals = [min(sample) - 0.1 + i * 0.01 for i in range(int((max(sample) - min(sample) + 0.2) / 0.01) + 1)]
teoretical_y = [normal_cdf(x, mean, sigma) for x in x_vals]

# Построение графиков
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))

# Эмпирическая функция распределения
plt.step(empirical_x, empirical_y, where='post', label='Эмпирическая функция', color='blue')

# Теоретическая функция распределения
plt.plot(x_vals, teoretical_y, label='Теоретическая функция (нормальная)', color='red', linestyle='--')

# Настройка графика
plt.title('Эмпирическая и теоретическая функции распределения')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.grid(True)
plt.legend()
plt.show()
```

$\overline{\sigma^2}=\frac{1}{n}\sum_{i}^n(x_{i}-a)^2=\frac{1}{12}((7.95-8.2)^2+(8.02-2)^2+\dots)$  
$\overline{\sigma^2}=\frac{1}{12}(0.0625+\dots+0.0484)=\frac{0.2342}{12}=0.0195$

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, uniform

# Данные
n = 100  # общее число наблюдений
l = 12  # количество интервалов
x_min = -6  # минимальное значение
x_max = 6   # максимальное значение
n_i = [2, 3, 6, 10, 12, 16, 15, 14, 11, 5, 4, 2]  # частоты

# Ширина интервала
h = (x_max - x_min) / l

# Границы интервалов
intervals = [(x_min + i * h, x_min + (i + 1) * h) for i in range(l)]

# Средние значения интервалов
x_i_mid = [(interval[0] + interval[1]) / 2 for interval in intervals]

# Эмпирические частоты
p_i = [ni / n for ni in n_i]

# Относительные частоты
p_i_relative = [pi / h for pi in p_i]

# Построение гистограммы
plt.figure(figsize=(10, 6))
plt.bar(x_i_mid, p_i_relative, width=h, edgecolor='black', align='center', color='skyblue', label='Гистограмма частот')

# Оценка для нормального распределения
mean_x = sum(x * n for x, n in zip(x_i_mid, n_i)) / sum(n_i)  # Среднее
std_x = np.sqrt(sum(n * (x - mean_x) ** 2 for x, n in zip(x_i_mid, n_i)) / sum(n_i))  # Стандартное отклонение

# Нормальное распределение
xmin, xmax = plt.xlim()
x_vals = np.linspace(xmin, xmax, 100)
y_vals_normal = norm.pdf(x_vals, mean_x, std_x)
plt.plot(x_vals, y_vals_normal, 'r-', label=f'Нормальное: μ={mean_x:.2f}, σ={std_x:.2f}')

# Равномерное распределение с постоянной плотностью 0.8
density_uniform = 0.08
range_uniform = 1 / density_uniform  # Длина интервала равномерного распределения
a_uniform = mean_x - range_uniform / 2
b_uniform = mean_x + range_uniform / 2
y_vals_uniform = [density_uniform] * len(x_vals)  # Плотность постоянна на всем интервале
plt.plot(x_vals, y_vals_uniform, 'g-', label=f'Равномерное: a={a_uniform:.2f}, b={b_uniform:.2f}')

# Настройки графика
plt.title('Гистограмма частот и оценки распределений', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('Плотность вероятности', fontsize=12)
plt.xticks(np.arange(x_min, x_max + h, h), rotation=45)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend()

plt.show()
```
